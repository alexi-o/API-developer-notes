#### 1.2.  Loading the data

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

**2.3. Cleaning the data (part 2)**

```python
# Pandas supplies melt(), which allows a user to melt values along the column headings of a Dataframe into a single column of data
clean_properties = properties_transposed.melt(id_vars=['London_Borough', 'ID'])
```

```python
# Checking data types to ensure they are all correct
clean_properties.dtypes
```

```python
clean_properties.dtypes
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
# Checking to see if there are any missing values in the current dataframe
clean_properties.count()
```

```python
# Looks like the counts across the columns do NOT align meaning some of the columns
# are missing values
unique_boroughs = clean_properties['London_Borough'].unique()
```

```python
# Seems that we have an access of London Boroughs in our dataframe
# The len() method should return 32, not 48, meaning an excess
# of 16 boroughs present.
unique_boroughs
```

```python
# Seems that these boroughs contain a lot of NaN values so it's pruning time
NaNFreeDF = clean_properties[clean_properties['Average_price'].notna()]
NaNFreeDF['London_Borough'].unique()
```

```python
# The value counts in each column now match! Now we need to remove the remaining
# excess boroughs (comparing against those found on our source list: https://en.wikipedia.org/wiki/London_boroughs#Former_authorities)
nonBoroughs = ['Inner London', 'Outer London', 
               'NORTH EAST', 'NORTH WEST', 'YORKS & THE HUMBER', 
               'EAST MIDLANDS', 'WEST MIDLANDS',
              'EAST OF ENGLAND', 'LONDON', 'SOUTH EAST', 
              'SOUTH WEST', 'England']
```

```python
# We can use this list in a filter statement with a negation operator (~) to filter out those nonBoroughs from out dataframe
df = NaNFreeDF[~NaNFreeDF['London_Borough'].isin(nonBoroughs)]
df['London_Borough'].unique()
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# We need to have the mean house price for each individual year
dfg = df.groupby(by=['London_Borough', 'Year']).mean()
dfg.sample(5)
```

```python
dfg = dfg.reset_index()
```

```python
def price_ratio(borough):
  year1998 = float(borough['Average_price'][borough['Year'] == 1998])
  year2018 = float(borough['Average_price'][borough['Year'] == 2018])
  ratio = [year1998 / year2018]
  return ratio
```

```python
price_ratio(dfg[dfg['London_Borough'] =='City of London'])
```

```python
# Creating an empty dictionary to store ratio values for each borough
final = {}

for b in dfg['London_Borough'].unique():
  borough = dfg[dfg['London_Borough'] == b]
  final[b] = price_ratio(borough)

print(final)
```

```python
# Creating a dataframe with the new data
df_ratios = pd.DataFrame(final)
df_ratios.head()
```

```python
# The data needs to be transposed to get the column headings to show as rows in the index column
df_ratios_transposed = df_ratios.T
df_ratios_transposed = df_ratios_transposed.reset_index()
df_ratios_transposed.head()
```

```python
# Let's slap some more lipstick on that pig
df_ratios_transposed.rename(columns={"index": "London_Borough", 0: "2018"})
df_ratios_transposed.head()
```

```python
# Extracting and printing the boroughs with the 15 highest changes in price
top15 = df_ratios_transposed.sort_values(by='2018', ascending=False).head(15)
print(top15)
```

```python
# Plotting the data
ax = top15[['London_Borough', '2018']].plot(kind='bar')

ax.set_xticklabels(top15.London_Borough)
```

```python
# In looking at the information about the boroughs in https://en.wikipedia.org/wiki/London_boroughs#Former_authorities
# I noticed the Designation column of the boroguh data which corresponds to the relative location of the particular borough
# either equaling Inner or Outer.
# This data could be used to draw additional conclusions about the relative trends 
designations = {
    'Camden': 'inner',
    'Greenwich': 'inner',
    'Hackney': 'inner',
    'Hammersmith & Fulham': 'inner',
    'Islington': 'inner',
    'Kensington & Chelsea': 'inner',
    'Lambeth': 'inner',
    'Lewisham': 'inner',
    'Southwark': 'inner',
    'Tower Hamlets': 'inner',
    'Wandsworth': 'inner',
    'Westminster': 'inner',
    'Barking & Dagenham': 'outer',
    'Barnet': 'outer',
    'Bexley': 'outer',
    'Brent': 'outer',
    'Bromley': 'outer',
    'Croydon': 'outer',
    'Ealing': 'outer',
    'Enfield': 'outer',
    'Haringey': 'outer',
    'Harrow': 'outer',
    'Havering': 'outer',
    'Hillingdon': 'outer',
    'Hounslow': 'outer',
    'Kingston upon Thames': 'outer',
    'Merton': 'outer',
    'Newham': 'outer',
    'Redbridge': 'outer',
    'Richmond upon Thames': 'outer',
    'Sutton': 'outer',
    'Waltham Forest': 'outer'
}
```

```python
# Creating a new column 'Designation' with respective values for each borough
df_ratios['Designation'] = df_ratios['Borough'].map(designations)
```

```python
inner_boroughs = df_ratios[df_ratios['Designation'] == 'inner']
outer_boroughs = df_ratios[df_ratios['Designation'] == 'outer']
```

```python
def average_ratio(data):
  sum = 0
  for d in data:
    sum += d

  return [sum / len(data)]
```

```python
print("Average price change ratio for inner boroughs:", average_ratio(inner_boroughs['2018']))
print("Average price change ratio for outer boroughs:", average_ratio(outer_boroughs['2018']))
```
